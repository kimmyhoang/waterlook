```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your dataset is in a CSV file
df = pd.read_csv('soil_data.csv')

# Convert the 'status' column from 'on'/'off' to 1/0
df['Status'] = df['Status'].map({'ON': 1, 'OFF': 0})

# Exclude rows where Time is between 90 and 110
df_filtered = df[~df['Time'].between(70, 110)]

# Calculate the correlation matrix including the 'status' column after filtering
correlation_matrix = df_filtered.corr()

# Plotting the updated correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix (Excluding Time 90-110)")
plt.show()

# If you want to specifically visualize the correlation of 'Time' with other columns
time_corr = correlation_matrix['Time'].drop('Time')  # Exclude time from the correlation with itself
time_corr_sorted = time_corr.sort_values(ascending=False)

# Plotting the specific correlation for 'Time' with other columns
plt.figure(figsize=(8, 6))
time_corr_sorted.plot(kind='bar', color='skyblue')
plt.title("Correlation of Time with Other Columns (Excluding Time 90-110)")
plt.xlabel("Columns")
plt.ylabel("Correlation Coefficient")
plt.show()
```
